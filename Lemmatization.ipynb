```python
file = open('deceptive_preprocessed.txt')
```

```python
import spacy

nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
```

```python
decept = []
for line in file:
    data_line = line.rstrip().split('.  ')
    decept.extend(data_line)


decept
```

```python
decept_lemma = []    
for sentence in decept:
    dog = [token.lemma_ for token in nlp(sentence)]
    decept_lemma.append(' '.join(dog))
decept_lemma    
```

```python
import pandas as pd    

df = pd.DataFrame(decept_lemma)
df.to_csv('decept_lemma.csv', index=False)
```

```python
file2 = open('truthful_preprocessed.txt')
```

```python
truth = []
for line in file2:
    data_line = line.rstrip().split('. ')
    truth.extend(data_line)
```

```python
truth_lemma = []    
for sentence in truth:
    dog = [token.lemma_ for token in nlp(sentence)]
    truth_lemma.append(' '.join(dog))
truth_lemma    
df = pd.DataFrame(truth_lemma)
df.to_csv('truth_lemma.csv', index=False)
```
